```python
# Import the yfinance library
import yfinance as yf

# Create a Ticker object for Tesla (symbol: TSLA)
tesla = yf.Ticker("TSLA")

# Extract historical stock data for Tesla (default period is max)
tesla_data = tesla.history(period="max")

# Reset the index to make 'Date' a column
tesla_data.reset_index(inplace=True)

# Save the dataframe to a CSV file in Watson Studio project storage
tesla_data.to_csv("tesla_data.csv", index=False)

# Display the first five rows
tesla_data.head()
```

```python
# Install required libraries (run once)
!pip install requests beautifulsoup4 pandas

# Import necessary libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL for Tesla revenue data
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# Set a user-agent to mimic a browser request
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"}

# Download the webpage
response = requests.get(url, headers=headers)

# Check if request was successful
if response.status_code == 200:
    html_data = response.text
else:
    print(f"Failed to retrieve page. Status code: {response.status_code}")
    html_data = ""

# Parse the HTML data
soup = BeautifulSoup(html_data, "html.parser")

# Create an empty DataFrame for Tesla revenue
tesla_revenue = pd.DataFrame(columns=["Date", "Revenue"])

# Find the table with Tesla Quarterly Revenue
tables = soup.find_all("table")
for table in tables:
    if "Tesla Quarterly Revenue" in str(table):
        rows = table.find("tbody").find_all("tr")
        for row in rows:
            cols = row.find_all("td")
            if len(cols) == 2:
                date = cols[0].text
                revenue = cols[1].text.replace("$", "").replace(",", "")
                tesla_revenue = pd.concat([tesla_revenue, pd.DataFrame({"Date": [date], "Revenue": [revenue]})], ignore_index=True)

# Remove empty revenue entries
tesla_revenue = tesla_revenue[tesla_revenue["Revenue"] != ""]

# Display the last five rows
tesla_revenue.tail()
```

```python
# Import libraries
import yfinance as yf
import plotly.graph_objects as go
from plotly.offline import init_notebook_mode

# Initialize plotly for notebook mode
init_notebook_mode(connected=True)

# Define the make_graph function
def make_graph(data, title):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=data['Date'], y=data['Close'], mode='lines', name='Close Price'))
    fig.update_layout(
        title=title,
        xaxis_title='Date',
        yaxis_title='Stock Price (USD)',
        template='plotly_white'
    )
    fig.show(renderer="iframe")

# Create a Ticker object for Tesla (symbol: TSLA)
tesla = yf.Ticker("TSLA")

# Extract historical stock data for Tesla
tesla_data = tesla.history(period="max")

# Reset the index to make 'Date' a column
tesla_data.reset_index(inplace=True)

# Debug: Check if data is loaded
print("Number of rows in tesla_data:", len(tesla_data))
print("Sample data (first row):")
print(tesla_data.head(1))

# Plot the Tesla stock data with a title
make_graph(tesla_data, "Tesla Stock Price Over Time")
```
